```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
df = pd.read_csv("C:\\Users\\Amisha\\stock_analysis\\csv_data\\indexProcessed.csv")
df.head()
```

```python
df_info = pd.read_csv("C:\\Users\\Amisha\\stock_analysis\\csv_data\\indexInfo.csv")
df_info.head()
```

```python
df_merge = df.merge(df_info, on = "Index", how = "inner" )
df_merge
```

```python
df_merge.info()
```

```python
df_merge['Date'] = pd.to_datetime(df_merge['Date'])
```

```python
df_merge.info()
```

```python
df.Index.value_counts()
```

```python
options = ["Europe","Switzerland"]
df_se = df_merge[df_merge["Region"].isin(options)]
df_se.head()
```

```python
df_se.Index.value_counts()
```

```python
import plotly.express as px
%matplotlib inline
```

```python
fig = px.line(df_se, x = 'Date',y = 'Open',color = 'Region',labels = {"Index" : "Region"},
              color_discrete_map = {"Europe":"#4363d8","Switzerland":"#e6194B",})

fig.update_layout(
    plot_bgcolor = "White",
    title_text = "Open stock price for Swiss and Europe Exchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black")
fig.show()
```

```python
fig = px.line(df_se, x = 'Date',y = 'Close',color = 'Region',labels = {"Index" : "Region"},
              color_discrete_map = {"Europe":"#4363d8","Switzerland":"#e6194B",})

fig.update_layout(
    plot_bgcolor = "White",
    title_text = "Open stock price for Swiss and Europe Exchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black")
fig.show()
```
